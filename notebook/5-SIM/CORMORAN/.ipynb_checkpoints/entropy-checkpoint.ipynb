```python
from pylayers.measures.cormoran import *
import pylayers.simul.simultraj as st
import h5py
from pylayers.util.CDF import *
%matplotlib inline
```

```python
S=  CorSer(serie =6, day=11)
```

```python
serie  = 's'+str(S.serie)
print serie
```

```python
i0 = 2
link_list = list(S.hkb.keys())
time = array(S.hkb.index)[i0:]
```

```python
dev_arr = np.array(map( lambda x : x.split('-'), link_list))
```

```python
body_link = filter( lambda x : x.split('-')[0][0:2] != 'AP'  and x.split('-')[1][0:2] != 'AP'  , link_list)
indx1 = map(lambda x : link_list.index(x) , body_link)
```

```python
body_link
```

```python
b2b_link =filter( lambda x : x.split('-')[0].split(':')[0] != x.split('-')[1].split(':')[0]  ,body_link)
indx2 = map(lambda x : link_list.index(x) ,b2b_link)
#b2b_link
```

```python
ob_link =filter( lambda x : x.split('-')[0].split(':')[0] == x.split('-')[1].split(':')[0]  ,body_link)
indx3 = map(lambda x : link_list.index(x) ,ob_link)
ob_link
```

```python
ltyp = 'body'
if ltyp == 'all':
    indx = link_list.index
    llinks  = link_list
elif ltyp == 'body':
    indx = indx1
    llinks = body_link
elif ltyp == 'b2b':
    indx = indx2
    llinks = b2b_link
elif ltyp == 'ob':
    indx = indx3 
    llinks = ob_link
        
data = S.hkb.values[i0:,indx]
data.shape
```

```python
plt.figure(figsize=(12,10))
plt.imshow(data.T, interpolation = 'None')
plt.axis('tight')
plt.colorbar()
```

```python
entropy=[]
nbin=50
for ilink in range(len(llinks)):
    D=data.T[ilink,:]
    H=plt.hist(D,nbin)
    p=H[0]
    p = p/sum(p)
    x=H[1]
    e = -p*np.log(p)/np.log(2)
    ep = np.nan_to_num(e)
    s = sum(ep)*(x[1]-x[0])
    entropy.append(s)
```

```python
plt.plot(entropy)
plt.grid()
plt.xlabel('link index')
plt.ylabel('Entropy')
```

```python
u = np.argsort(entropy)
entropy= np.array(entropy)
plt.plot(entropy[u])
plt.grid()
```

```python
fig, ax = plt.subplots(figsize = (35,20))
ctyp = 'Entropy' # Std or Variance or Corrr
ind = u
infos = entropy 
im = ax.imshow(data.T[ind,], interpolation = 'None')
plt.axis('tight')

tick_lbls = time[::1000]
tick_locs = np.arange(0,len(time))[::1000]
plt.xticks(tick_locs, tick_lbls.astype(int), fontsize  =20)
plt.yticks(range(len(llinks)), array(llinks)[ind], fontsize = 14)
divider = make_axes_locatable(ax)
plt.xlabel('time (s)',fontsize = 25)
cax = divider.append_axes("right", size="5%", pad=0.05)
cbar = plt.colorbar(im, cax=cax )
axy = divider.append_axes("right", 2, pad=0.5, sharey=ax)
axy.plot(infos[ind],range(len(llinks)),'o--')
plt.grid()
plt.setp( axy.get_yticklabels(), visible=False)
plt.xlabel(ctyp, fontsize = 20)
axy.tick_params(axis='x', labelsize=10)
plt.savefig("/private/staff/i/bi/mmhedhbi/Figures/"+serie+"_links_"+str(ltyp)+"_"+str(ctyp)+".jpg")
```
