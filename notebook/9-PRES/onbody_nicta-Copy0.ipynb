## GENERATE CIR for the link 

```python
from pylayers.mobility.ban.body import *
import pylayers.simul.simultraj as st
from pylayers.mobility.trajectory import *
from pylayers.simul.link import *
import pylayers.signal.bsignal as bs

L=Layout('nicta.ini')
B=Body('Bernard2.ini')
T=Trajectory()

t=np.arange(0,10,0.1)
pt=np.zeros((len(t),3))
pt[:,0]=np.arange(0,0.1,0.001)+1.5
pt[:,1]=np.arange(0,10,0.1)+4.25
pt[:,2]=1.5
T.generate(t=t,pt=pt,ID='0',name='Bernard2.ini',typ='ag')

DL=DLink(L=L,verbose=True)
la=[]
taumin = 0
taumax = 300
taustep = 0.1
CIR = []
x = np.arange(taumin,taumax,taustep)
y = np.zeros(len(x))
cira=bs.TUsignal(x,y)
for tt in t:
    if tt < 9.8:
        print tt
        B.settopos(T,t=tt,cs=True,treadmill=True,p0=pt[0,:2])
        DL.a=B.dcs['0'][:,0]
        DL.Ta = B.acs['0']
        DL.b=B.dcs['1'][:,0]
        DL.Tb= B.acs['1']
        DL.eval()
        cira=bs.TUsignal(x,y)
        cira.aggcir(DL.H.ak,DL.H.tk)
        CIR.append(cira)
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
for c in CIR:
    pl = 10*np.log10(np.sum(abs(c.y)**2, axis= 1))[3:]
    #pl = 10*np.log10((abs(CIR[i].y)**2))
    print t.shape
    print pl.shape
    ax.plot(t,pl,'r:',linewidth=2,)

    xlabel('Time (s)')
    ylabel('Path Loss (dB)')
    ax.axis((0,5,-80,-30))
plt.tight_layout()    

plt.show()
```
